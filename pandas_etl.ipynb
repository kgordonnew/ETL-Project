### Create database connection

```python
state_df = pd.read_csv("States.csv")
state_df.head()
```

```python
death_causes_df.dtypes
```

```python
#join to columns
death1_causes_df=pd.merge(death_causes_df, state_df, on = "State")
death1_causes_df.head()
```

```python
#outer: use union of keys from both frames, similar to a SQL full outer
#      join; sort keys lexicographically
dfmerge=pd.merge(death1_causes_df, city_cols,  how='outer', on=['Year','StateAbbr'], validate="m:m")
```
