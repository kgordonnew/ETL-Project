## Database

```python
# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import create_engine
Base = declarative_base()
# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float
# Import the session object so that we can begin to push the toxic release data to the database
from sqlalchemy.orm import Session
```

```python
# Create class for toxic release data

class airQualityIndex(Base):
    __tablename__ = 'AirQualityIndex'
    id = Column(Integer, primary_key=True)
    state = Column(String(255))
    aqi = Column(Float)
```

```python
engine = create_engine("sqlite:///etlproject.sqlite")
conn = engine.connect()

# Create the ToxicRelease table within the database
Base.metadata.create_all(conn)

session = Session(bind=engine)
```

```python
air_state = air_data.groupby('State')['90th Percentile AQI'].mean()
air_state = pd.DataFrame(air_state)
air_state.head()
```
